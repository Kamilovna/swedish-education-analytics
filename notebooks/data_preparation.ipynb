# Data Sources and Selected Variables

For each year, data was retrieved from table 3. For the years 2020–2022, the data was supplemented with information from table 4, where the variables total sought places and total granted places were summed per case number before being joined with the main table.

Additionally, all Swedish column names were renamed to English during data processing to improve readability and consistency throughout the analysis.

### Variables Used in the Analysis

### The following variables are available for all years:

- *diarienummer* - case_number
- *utbildningsnamn* - education_name
- *utbildningsområde* - education_area
- *beslut* - decision
- *kommun* - municipality
- *län* - county
- *yh_poäng* - education_credits
- *studieform* - study_mode
- *studietakt_procent* - study_pace_percentage
- *utbildningsanordnare* - education_provider
- *huvudmannatyp* - principal_type

### The following two variables were retrieved from table 4 for the years 2020–2022 and from table 3 for the remaining years:

- *sökta_platser_totalt* - total_sought_places
- *beviljade_platser_totalt* - total_granted_places

### The following variables are only available for the years 2023–2025:

- *sun5_inriktning* - sun5_code
- *sun5_inriktning_namn* - sun5_name
- *seqf_nivå* - seqf_level
- *smalt_yrkesområde* - narrow_occupational_area

### The following variables were created during data processing for all years:

- source_year
- source_file
- source_sheet

The variables were selected to enable analyses of the development of the education supply over time, the geographical distribution of educations, different education areas and providers, as well as the relationship between the number of sought and granted education places.

## Normalization of the variable beslut (decision)
Kolumnen beslut normaliserades för att säkerställa en enhetlig kategorisering över hela tidsperioden. I data för åren 2020–2021 användes värdena “beviljad” och “ej beviljad”, medan senare år innehöll kategorierna “beviljad” och “Avslag”. För att skapa en konsekvent variabel ersattes värdet “ej beviljad” med “Avslag”. Efter normaliseringen innehåller kolumnen endast två kategorier: “Beviljad” och “Avslag”.


```python
resultat = {
    2020: resultat_2020,
    2021: resultat_2021,
    2022: resultat_2022,
    2023: resultat_2023,
    2024: resultat_2024,
    2025: resultat_2025
}
for year, df in resultat.items():
    print(year)
    print(sorted(df["Beslut"].unique()))
    print()
```

```python
resultat_2020["Beslut"] = resultat_2020["Beslut"].replace(
    {"Ej beviljad": "Avslag"}
)
resultat_2021["Beslut"] = resultat_2021["Beslut"].replace(
    {"Ej beviljad": "Avslag"}
)
```

```python
rename_dict = {
    "Diarienummer": "case_number",
    "Utbildningsnamn": "education_name",
    "Utbildningsområde": "education_area",
    "Beslut": "decision",
    "Kommun": "municipality",
    "Län": "county",
    "YH-poäng": "education_credits",
    "Studieform": "study_mode",
    "Studietakt %": "study_pace_percentage",
    "Utbildningsanordnare administrativ enhet": "education_provider",
    "Huvudmannatyp": "principal_type",
    "SUN5 inriktning": "sun5_code",
    "SUN5 inriktning namn": "sun5_name",
    "SeQF nivå": "seqf_level",
    "Smalt yrkesområde": "narrow_occupational_area",
    "Sökta platser totalt": "total_sought_places",
    "Beviljade platser totalt": "total_granted_places"
}
```

### Create column structure

A list of all variables to be included in the final analysis was created (main_columns). Since the variables sun5_inriktning, sun5_inriktning_namn, seqf_nivå and smalt_yrkesområde only exist for the years 2023–2025, reindex() was used to give all DataFrames the same column structure before merging the data.
If a variable was missing for a certain year, the column was automatically created with empty values (NaN). This facilitated the subsequent merging of data from different years.

```python
main_columns = [
    "source_year",
    "source_file",
    "source_sheet",
    "case_number",
    "education_name",
    "education_area",
    "decision",
    "municipality",
    "county",
    "education_credits",
    "study_mode",
    "study_pace_percentage",
    "education_provider",
    "principal_type",
    "sun5_code",
    "sun5_name",
    "seqf_level",
    "narrow_occupational_area",
    "total_sought_places",
    "total_granted_places"
]
```

```python
applications_2020 = resultat_2020.reindex(columns=main_columns)
applications_2021 = resultat_2021.reindex(columns=main_columns)
applications_2022 = resultat_2022.reindex(columns=main_columns)
applications_2023 = resultat_2023[main_columns]
applications_2024 = resultat_2024[main_columns]
applications_2025 = resultat_2025[main_columns]
```

The variable principal_type contained the categories Privat, Kommun, Landsting, Statlig och Region. This resulted in a more consistent classification in the dataset.
